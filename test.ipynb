```python
import os
import sys
import torch
import pandas as pd
import numpy as np
from torch import nn
from torchvision import transforms
from torch.utils.data import Dataset
from PIL import Image
import matplotlib.pyplot as plt
```

```python
x = np.arange(-360,360,1)
y = 0
x = x * (torch.pi / 180)
y = y * (torch.pi / 180)
scale = 1

pred = torch.tensor(x)
target = torch.tensor(y)
err = torch.abs(pred-target)
loss = ((torch.remainder((err + torch.pi), 2*torch.pi) - torch.pi)**2)
baseL =(((torch.cos(pred) - torch.cos(target))*scale)**2 + ((torch.sin(pred) - torch.sin(target))*scale)**2)
bladeL = (((torch.cos(pred*3) - torch.cos(target*3))*scale)**2 + ((torch.sin(pred*3) - torch.sin(target*3))*scale)**2)*1/4
L = baseL + bladeL
```

```python
plt.plot(x,loss)
plt.show()

plt.plot(x,baseL)
plt.plot(x,bladeL)
plt.plot(x,L)
plt.show()
```
